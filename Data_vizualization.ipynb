```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('nz_weather.csv')
print(data.head(13))
```

```python
data.dtypes
```

```python
var = data.sort_values(by='DATE' , ascending=False)
print(var)
```

```python
print(var.head(3))
```

```python
x = data.DATE
y = data.Auckland
y_1 = data.Christchurch
plt.xlabel('Data' , fontsize = 12)
plt.ylabel('Temperature' , fontsize = 14)
plt.title('Temperature in Auckland and Christchurch')
plt.scatter(x,y )
plt.scatter(x ,y_1) 
plt.xticks(rotation = 90)
plt.show()
```

```python
yy = data.head(15)
plt.figure(figsize=(20,10))
y = yy.Auckland
y_1 = yy.Dunedin
x = yy.DATE
plt.plot(x,y ,label='Auckland')
plt.plot(x ,y_1 , label = 'Dunedin')
plt.xlabel('Date' , fontsize=12 ,rotation=0)
plt.ylabel('Temperature' , fontsize=12)
plt.title('Line Plot' , fontsize=45)
plt.legend()
```

```python
yy = data
y = yy.Auckland
plt.title('Histogram' , fontsize=45)
plt.xlabel('Temperature',fontsize=12)
plt.ylabel('No.of Days',fontsize=12)
plt.hist(y,bins=20)
```

```python
obj = ['Tavolina' , 'Karrige' , 'Tabela' , 'Dritare']
counts = [15 , 30 , 1 , 4]
plt.title('Bar Chart',fontsize=45)
plt.bar(obj , counts)
```

```python
box_plot = data.Auckland
plt.title('Boxplot' ,fontsize=45)
plt.boxplot(box_plot)
```

```python
x1 = yy.DATE
y1 = yy.Auckland
y2 = yy.Dunedin
a,(g1 , g2) = plt.subplots(1  , 2,sharey=True)
a.suptitle('Sharing Y')
g1.plot(x1,y1)
g2.plot(x1,y2)
```

```python
x = data.DATE
y = data.Auckland
fig , G1 = plt.subplots(2 , sharex=True)
G1[0].plot(x , y)
G1[1].plot(x , y)
```

```python
data1 = data.DATE
data2 = data.Auckland
data3 = data.Hamilton
plt.figure(figsize=(20,20))
plt.plot(data1,data2 , label='Auckland' )
plt.plot(data1,data3,label='Hamilton')
plt.title('Weather in Hamilton nd Auckland')
plt.xticks(rotation=90 )
plt.legend()

```
